```python
import numpy as np

array1=np.array([10,20,30])

array2=np.array([2,2,2])

print("array2 multiplied by array1: ", array1*array2)

print("array2 divided by array1: ", array2/array1)

print("array2 raised to the power of array1: ", array2**array1)

print("Adding two numpy arrays {array} and {array} together:", array1+array2)
```

```python
import numpy as np
array1=np.array([10,20,30])

# sine function print("Sine: ",np.sin(array1))

# logarithm

print("Natural logarithm: ", np.log(array1))

print("Base-10 logarithm: ", np.log10(array1))

print("Base-2 logarithm: ",np.log2(array1))

# Exponential

print("Exponential: ",np.exp(array1))
```

```python
# Sample datasets
dataset1 = [1, 2, 3, 4]
dataset2 = [5, 6, 7, 8]

# Multiplication
result_multiplication = [a * b for a, b in zip(dataset1, dataset2)]
print("Multiplication Result:", result_multiplication)

# Division
result_division = [a / b for a, b in zip(dataset1, dataset2)]
print("Division Result:", result_division)

# Power (exponentiation)
power_exponent = [a ** b for a, b in zip(dataset1, dataset2)]
print("Power Result:", power_exponent)
```

```python
import numpy as np


data = np.array([1, 2, 3, 4])


sin_result = np.sin(data)
print("Sin Result:", sin_result)


log_result = np.log(data)
print("Natural Log Result:", log_result)


log2_result = np.log2(data)
print("Log Base 2 Result:", log2_result)


exp_result = np.exp(data)
print("Exponential Result:", exp_result)
```

```python
people = ['Arun', 'Banu', 'Chendhu', 'David', 'Eshwar', 'harish', 'Gagan', 'Hamish','Indra', 'Jagan', 'Kumar', 'Loki']

age = [21,12,32,45,37,18,28,52,5,40,48,15]

weight = [55,35,77,68,70,60,72,69,18,65,82,48]

height = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter (weight, height)

plt.title("Relationship between Weight and Height of Patients")

plt.ylabel("Height in Meters")

plt.xlabel("Weight of Patients in Kgs")

plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')


plt.bar(x=people,height=weight, width=0.6,
        color='orange',edgecolor='k',alpha=0.6)

plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```
